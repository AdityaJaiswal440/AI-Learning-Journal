```python
# Stemming
```

```python
words = ["eating", "eats", "eaten", "connection", "connect", "connector", "connected", "connecting", "faster", "fasting", "fasten", "faster", "fast"]
```

```python
#PorterStemmer
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('walking')
```

```python
reg_stemmer.stem('fasts')
```

```python
# Snowball Stemming
```

```python
from nltk.stem import SnowballStemmer
```

```python
sb_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word +"---->" + sb_stemmer.stem(word))
```

```python
stemming.stem('fairly'), stemming.stem('sportingly')
```

```python
sb_stemmer.stem('fairly'), sb_stemmer.stem('sportingly')
```
