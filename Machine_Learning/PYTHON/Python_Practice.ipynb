```python
students = [
    {"name": "Aditya", "score": 85},
    {"name": "Rahul", "score": 30},
    {"name": "Sneha", "score": 95}
]

class GradeAnalyzer:
    def __init__(self, student_list):
        self.students = students

    def show_passing(self):
        print("--- Passing Students ---")
        for s in self.students:
            if s["score"] > 40:
                print(f"{s['name']} Passed")
            else : 
                print(f"{s['name']} Failed")

    def average_score(self):
        total = 0
        count = 0
        for s in self.students:
            total += s["score"]
            count += 1
        
        avg = total / count
        print(f"Class Average: {avg}")

analyzer = GradeAnalyzer(students)


analyzer.show_passing()
analyzer.average_score()
```

```python
def is_strong_pass(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    return True

print(is_strong_pass("ADi12345"))
print(is_strong_pass("adi@12345"))
print(is_strong_pass("Adi@12345"))
```

```python
class BiasedDatasetError(Exception):
    """Raised when the dataset distribution is too skewed for training."""
    pass

def check_data_quality(labels):
    if len(set(labels)) < 2:
        raise BiasedDatasetError("Dataset must have at least 2 classes to train!")

try:
    my_labels = [1, 1, 1, 1]
    check_data_quality(my_labels)
except BiasedDatasetError as e:
    print(f"Critical Training Halt: {e}")
```

```python
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self,amount):
        self.balance += amount

    def withdraw(self,amount):
        if amount>self.balance:
            return "Insufficient Funds !!"
        else:
            print(f"Successfull withdrawl of {amount}")
            self.balance -= amount
        
    def get_balance(self):
        return self.balance

Cust_1 = BankAccount("Adi",100)
print(Cust_1.balance)
```

```python
Cust_1.deposit(1000)
```

```python
Cust_1.get_balance()
```

```python
Cust_1.withdraw(2100)
```
