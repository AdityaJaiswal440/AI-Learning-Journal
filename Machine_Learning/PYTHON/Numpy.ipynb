Slicing and Indexing
slicing: arr_name[row_start:row_end, col_start:col_end]

```python
import numpy as np

arr=np.random.randint(1,64,size=(1,8))
print(arr)
print(arr[:,2:6])
matrix = np.random.randint(4,98,size=(3,3))
print(matrix)
print(matrix[0:2,0:3])
```

Axis concept: 
axis = 0, Vertical
axis = 1, Horizontal

```python
matrix = np.array([
    [10, 20],  
    [30, 40]  
])

print(matrix.sum(axis=0)) 
print(matrix.sum(axis=1))
```

np.mean(): Average.

np.std(): Standard Deviation (How "spread out" the data is).

np.min() / np.max(): Lowest and Highest values.

np.argmax() / np.argmin(): Crucial. Returns the Index of the max/min value, not the value itself. (Used to find which class the model predicted with highest probability).

```python
scores = np.array([10, 50, 90, 20])

print(np.mean(scores))  
print(np.max(scores))    
print(np.argmax(scores)) 
```

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
vector = np.array([1, 0, 1])

result = matrix + vector 
# It effectively does:
# [1, 2, 3] + [1, 0, 1] = [2, 2, 4]
# [4, 5, 6] + [1, 0, 1] = [5, 5, 7]
print(result)
```

```python
marks = np.array([
    [50, 60, 70, 80], 
    [90, 95, 90, 95], 
    [30, 30, 30, 30]  
])
avg_subject = np.mean(marks,axis=0)
print(f"Subject Averages: {avg_subject}")

total_score = marks.sum(axis=1)
print(f"Total Scores: {total_score}")

top_index = np.argmax(total_score)
print(f"Index of Topper : {top_index}")
```
