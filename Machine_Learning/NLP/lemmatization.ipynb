```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer.lemmatize("caring")
```

```python
words = ["eating", "eats", "eaten", "connection", "connect", "connector", "connected", "connecting", "faster", "fasting", "fasten", "faster", "fast"]
```

```python
for word in words:
    print(word +"---->" + lemmatizer.lemmatize(word, pos="n"))
```
