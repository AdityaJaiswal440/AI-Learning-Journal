```python
import nltk
```

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize
```

```python
raw_text = "Hi!, I Aditya J, wants to check that how the sent_tokenize divides the string into tokens."
```

```python
sentences = sent_tokenize(raw_text)
```

```python
sentences
```

```python
type(sentences)
```

```python
for lines in sentences:
    print(lines)
```

```python
from nltk.tokenize import word_tokenize

for lines in sentences:
    print(word_tokenize(lines))
```

```python
from nltk.tokenize import TreebankWordDetokenizer
```

```python
tokenizer = TreebankWordDetokenizer()
```

```python
tokenizer.tokenize(raw_text)
```

```python
from nltk.tokenize import TreebankWordTokenizer

# 1. Instantiate the object
tokenizer = TreebankWordTokenizer() 

# 2. Execute the method
tokens = tokenizer.tokenize("I won't go.")
```

```python
print(tokens)
```

```python
from nltk.tokenize import word_tokenize

# Execute in one step
tokens2 = word_tokenize("I won't go.")
```
