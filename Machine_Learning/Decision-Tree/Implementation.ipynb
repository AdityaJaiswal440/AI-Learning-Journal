```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_iris  
```

```python
iris = load_iris()
```

```python
iris
```

```python
# based on 4 input features(Sepal length, sepal width, petal length,petal width), predict what type of iris flower is 
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
y=iris['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeClassifier= DecisionTreeClassifier(max_depth=3)
```

```python
treeClassifier.fit(X_train,y_train)
```

```python
# Visualizing the Decision tree
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(20,15))
tree.plot_tree(treeClassifier,filled=True)
```

```python
y_pred=treeClassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
# Pre-Pruning - for large datasets
```
