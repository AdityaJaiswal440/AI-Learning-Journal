```python
import numpy as np
```

```python
marks = np.random.randint(1,96,(20))
marks
```

```python
min,Q1,median,Q3,max = np.quantile(marks,[0,0.25,0.50,0.75,1.0])
```

```python
min,Q1,median,Q3,max
```

```python
IQR = Q3-Q1
IQR
```

```python
Lower_fence = Q1 - 1.5*(IQR)
Lower_fence
```

```python
Higher_fence = Q3 + 1.5*(IQR)
Higher_fence
```
